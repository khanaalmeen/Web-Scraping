```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphones"
```

```python
wanted_list=["Apple iPhone 13 (256GB) - Blue","₹80,990","6,563"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_8vi1':'Title','rule_oq0o':'Price'})
scraper.keep_rules(['rule_8vi1','rule_oq0o'])
scraper.save("amazon-search")
```

```python
results=scraper.get_result_similar("https://www.amazon.in/s?k=vivo",group_by_alias=True)
```
